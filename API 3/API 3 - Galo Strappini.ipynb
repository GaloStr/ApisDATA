```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
try:
    df = pd.read_csv('esperanza_de_vida.csv', encoding='latin1')
    print("Archivo CSV cargado correctamente.")
    print(df.head())
except FileNotFoundError:
    print("Error: No se encontró el archivo CSV. Verifica la ruta.")
except Exception as e:
    print(f"Ocurrió un error al cargar el archivo: {e}")
```

```python
year2025 = df[df['anio'] == 2025]
plt.style.use('dark_background')
plt.figure(figsize=(9,3))
plt.hist(year2025['mujeres'], bins=10 , edgecolor='black', color='pink')
plt.xlabel('Esperanza de Vida de Mujeres (años)')
plt.ylabel('Frecuencia (Número de Provincias)')
plt.title('Distribución de la Esperanza de Vida de Mujeres en 2015')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.style.use('dark_background')
plt.scatter(df['varones'], df['mujeres'], alpha=0.75)
plt.xlabel('Esperanza de vida (Varones)')
plt.ylabel('Esperanza de vida (Mujeres)')
plt.title('Relación entre Esperanza de Vida de Mujeres y Varones')
plt.grid(color='gray', alpha=0.7, linestyle='--')
plt.show()
```

```python
bsas = df[df['provincia'] == 'Buenos Aires']
plt.figure(figsize=(5,4))
plt.style.use('dark_background')
plt.plot(bsas['anio'], bsas['mujeres'], marker= 'o', label='mujeres')
plt.plot(bsas['anio'], bsas['varones'], marker= 'o', label='varones')
plt.xlabel('Año')
plt.ylabel('Esperanza de vida (años)')
plt.title('Esperanza de Vida en Buenos Aires (Varones y Mujeres)')
plt.legend()
plt.grid(color='gray', alpha=0.7, linestyle='--')
plt.show()
```
