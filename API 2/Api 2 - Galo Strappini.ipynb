```python
import pandas as pd
import numpy as np
import os
```

```python
poblacion = pd.read_csv('poblacion.csv', encoding='latin1')
esperanza = pd.read_csv('esperanza_de_vida.csv', encoding='latin1')
hogares = pd.read_csv('hogares_viviendas_superficie.csv', encoding='latin1')

display(poblacion)
display(esperanza)
display(hogares)
```

```python
#.HEAD - MOSTRAR PRIMER FILA DE LA TABLA POBLACION
#.TAIL - MOSTRAR ULTIMA FILA DE LA TABLA POBLACION
#.SHAPE - MOSTRAR TAMAÑO DE LA TABLA POBLACION
display(poblacion.head())
display(poblacion.tail())
display(poblacion.shape)
display(poblacion.dtypes)
```

```python
# - Esta función filtra el DataFrame 'poblacion' para excluir las
# filas donde la columna 'provincia' es igual a 'Total País'.
pobla_filtrado = poblacion[poblacion['provincia'] != 'Total País']
pobla_filtrado
```

```python
#.MERGE - UNIR TABLAS
# - Une la tabla de poblacion con la tabla de esperanza de vida
# - La columna de unión es 'provincia'
# - El tipo de unión es 'left' (izquierda)
# - La tabla resultante se llama 'poblac_superf'
# - Se utiliza la columna 'provincia' de la tabla de poblacion 
# y la columna 'provincia' de la tabla de esperanza de vida
poblac_superf = pd.merge(pobla_filtrado, hogares, left_on=['provincia'],
                        right_on=['provincia'], how='left')
poblac_superf
```

```python
# - Esta función calcula la densidad poblacional dividiendo la
# población total entre la superficie en km2
poblac_superf['densidad_poblacional'] = (poblac_superf['poblacion_total'] 
                                         / poblac_superf['superficie_km2'])
poblac_superf
```

```python
# - Muestra el número de valores nulos en cada columna de la tabla poblac_superf
poblac_superf.isnull().sum()
```

```python
poblac_superf['densidad_poblacional'].describe()
```

```python
# - La función 'np.mean' calcula la media de la columna 'densidad_poblacional'
# - La función 'np.std' calcula la desviación estándar de la columna 'densidad_poblacional'
# - Un valor de 'fuera_de_rango' mayor a 3 indica que el valor está muy por encima de la media
poblac_superf['fuera_de_rango'] = ((poblac_superf['densidad_poblacional'] - 
                                   np.mean(poblac_superf['densidad_poblacional'])) / np.std(poblac_superf['densidad_poblacional']))
poblac_superf

```

```python
poblac_superf[['provincia', 'densidad_poblacional']].drop_duplicates(subset=['provincia'])
```

```python
# - El formato especificado es '{:,.2f}', lo que significa que los números se mostrarán
# con dos decimales y con separadores de miles
pd.options.display.float_format = '{:,.2f}'.format
resumen_superficie = hogares['superficie_km2'].describe(percentiles=[0.25, 0.5, 0.75])

print("Medidas estadísticas de la superficie (en km²):")
print(resumen_superficie)
```
