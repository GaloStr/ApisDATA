```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("esperanza_de_vida.csv", encoding="latin1")
```

```python
# Hipótesis 1:
# "La esperanza de vida de las mujeres es sistemáticamente mayor que la de los varones en todas las provincias."
# Hipótesis 2:
# "La esperanza de vida ha aumentado con el tiempo en todas las provincias, para ambos géneros."

df['diferencia'] = df['mujeres'] - df['varones']

print("Promedio general de diferencia de esperanza de vida (mujeres - varones):")
print(df['diferencia'].mean())

avg_prov = df.groupby('provincia')[['mujeres', 'varones']].mean().sort_values(by='mujeres', ascending=False)
avg_prov.plot(kind='bar', color=['magenta', 'cyan'])
plt.title('Promedio de esperanza de vida por provincia (Mujeres vs Varones)')
plt.xlabel('Provincia')
plt.ylabel('Esperanza de vida promedio')
plt.xticks(rotation=90)
plt.legend(title='Género')
plt.tight_layout()
plt.show()

promedios_anuales = df.groupby('anio')[['mujeres', 'varones']].mean().reset_index()
promedios_anuales.plot(x='anio', y=['mujeres', 'varones'])
plt.style.use('dark_background')
plt.title('Promedio nacional de esperanza de vida por año')
plt.xlabel('Año')
plt.ylabel('Esperanza de vida')
plt.show()
```

```python
categorias = ['Ventas', 'Marketing', 'Producción', 'Ventas Online']
valores = [120, 90, 100, 80]

plt.bar(categorias, valores, color='skyblue')
plt.title('Ventas por Departamento', fontsize=16)
plt.xlabel('Departamento', fontsize=14)
plt.ylabel('Ventas (miles)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(valores):
    plt.text(i, v + 3, str(v), ha='center', fontsize=12)

plt.tight_layout()
plt.show()


meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas_A = [200, 210, 190, 220, 230, 240]
ventas_B = [180, 190, 200, 210, 205, 215]

plt.plot(meses, ventas_A, marker='o', label='Producto A', color='green')
plt.plot(meses, ventas_B, marker='o', label='Producto B', color='orange')

plt.title('Ventas Mensuales de Productos', fontsize=16)
plt.xlabel('Meses', fontsize=14)
plt.ylabel('Ventas (unidades)', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True)

plt.tight_layout()
plt.show()


# Datos ejemplo
etiquetas = ['Mujeres', 'Hombres']
porcentajes = [55, 45]
colores = ['#ff9999','#66b3ff']

plt.pie(porcentajes, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=90)
plt.title('Distribución por Género', fontsize=16)
plt.axis('equal')

plt.show()
```
