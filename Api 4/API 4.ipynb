```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("esperanza_de_vida.csv", encoding="latin1")
```

```python
# Hipótesis 1:
# "La esperanza de vida de las mujeres es sistemáticamente mayor que la de los varones en todas las provincias."
# Hipótesis 2:
# "La esperanza de vida ha aumentado con el tiempo en todas las provincias, para ambos géneros."


df['diferencia'] = df['mujeres'] - df['varones']

# Promedio general de diferencia
print("Promedio general de diferencia de esperanza de vida (mujeres - varones):")
print(df['diferencia'].mean())

avg_prov = df.groupby('provincia')[['mujeres', 'varones']].mean().sort_values(by='mujeres', ascending=False)
avg_prov.plot(kind='bar', color=['magenta', 'cyan'])
plt.title('Promedio de esperanza de vida por provincia (Mujeres vs Varones)')
plt.xlabel('Provincia')
plt.ylabel('Esperanza de vida promedio')
plt.xticks(rotation=90)
plt.legend(title='Género')
plt.tight_layout()
plt.show()

# Tendencia promedio por año
promedios_anuales = df.groupby('anio')[['mujeres', 'varones']].mean().reset_index()
promedios_anuales.plot(x='anio', y=['mujeres', 'varones'])
plt.style.use('dark_background')
plt.title('Promedio nacional de esperanza de vida por año')
plt.xlabel('Año')
plt.ylabel('Esperanza de vida')
plt.show()
```
